```python
from datasets import load_dataset, DatasetDict

# Tải dữ liệu
dataset = DatasetDict()

dataset['train'] = load_dataset("Thienpkae/Tele_vi", split='train')
dataset['validation'] = load_dataset("Thienpkae/Tele_vi", split='validation')
dataset['test'] = load_dataset("Thienpkae/Tele_vi", split='test')

# Kiểm tra lại tỷ lệ các phần
print(dataset)

# Lấy số lượng mẫu để kiểm tra
print("Train size:", len(dataset["train"]))
print("Validation size:", len(dataset["validation"]))
print("Test size:", len(dataset["test"]))
```

```python
dataset['train'] = dataset['train'].select_columns(['path', 'audio', 'sentence'])
dataset['validation'] = dataset['validation'].select_columns(['path', 'audio', 'sentence'])
dataset['test'] = dataset['test'].select_columns(['path', 'audio', 'sentence'])
```

```python
# Kiểm tra đường dẫn đến dataset
print("Cache directory for the dataset:", dataset.cache_files)

# In đường dẫn của từng tập dữ liệu
for split in dataset.keys():
    print(f"Path for {split} split:")
    for cache_file in dataset[split].cache_files:
        print(cache_file["filename"])
```

```python
from datasets import load_dataset, DatasetDict
import librosa
import soundfile as sf
import tqdm
dataset = DatasetDict()
dataset = load_dataset("Nguyen667201/test_data")
dataset
```

```python
def save_audio_file(array, path,sr = 16000):
    sf.write(path, array, sr)
```

```python
dataset["train"][3]["file"]
```

```python
import os
output = "/content/data"
if not os.path.exists(f"{output}/wavs"):
  os.makedirs(f"{output}/wavs")
with open(f'{output}/metadata.csv', 'w', encoding='utf8') as fp:
  print(f'file,text,duration', file=fp)
  for index, row in tqdm.tqdm(enumerate(dataset["train"])):
    text = row["text"]
    duration = row["duration"]
    path_wav_out = f"{output}/wavs/{index}.wav"
    save_audio_file(row["file"]["array"],path_wav_out)
    if os.path.exists(path_wav_out):
      print(f'{path_wav_out},{text},{duration}', file=fp)
    else
      print(f"khong co file nay {path_wav_out}")
```
